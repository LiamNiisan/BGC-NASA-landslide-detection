_This notebook shows how the pre training dataset used to train the multitask BERT model used for landslide event extraction was generated._

```python
squad = datasets.load_dataset("squad_v2")
```

```python
questions = squad["train"]["question"]
answers = squad["train"]["answers"]
contexts = squad["train"]["context"]
```

```python
where_questions = [(q, a, c) for q, a, c in zip(questions, answers, contexts) if "where" in q.lower()[:5]]
when_questions = [(q, a, c) for q, a, c in zip(questions, answers, contexts) if "when" in q.lower()[:5]]
```

```python
squad_extracted_data = []

for question in where_questions:
    entry = {
        "text" : question[2],
        "EVENT_CATEGORY": "",
        "EVENT_TRIGGER": "",
        "qas" : {
            "LOC" : question[1]["text"][0] if question[1]["text"] else "",
            "TIME" : "",
            "DATE" : "",
            "CAS" : ""
        }
    }
    squad_extracted_data.append(entry)


for question in when_questions:
    entry = {
        "text" : question[2],
        "EVENT_CATEGORY": "",
        "EVENT_TRIGGER": "",
        "qas" : {
            "LOC" : "",
            "TIME" : "",
            "DATE" : question[1]["text"][0] if question[1]["text"] else "",
            "CAS" : ""
        }
    }
    squad_extracted_data.append(entry)
```

```python
len(squad_extracted_data)
```

```python
import json

with open("../data/processed/squad_extracted_questions.json", "w") as f:
    json.dump(squad_extracted_data, f)
```
